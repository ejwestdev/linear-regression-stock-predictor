```python
import requests
import pandas as pd
from sklearn.linear_model import LinearRegression
from datetime import datetime
```

```python

api_url = "https://www.alphavantage.co/query"
api_key = "O1SJA6ND15QHG1W8"
params = {
    "function": "TIME_SERIES_DAILY_ADJUSTED",
    "symbol": "AMZN",
    "outputsize": "full",
    "apikey": api_key
}

response = requests.get(api_url, params=params)
data = response.json()
```

```python
# Extract the daily adjusted prices from the response
time_series = data["Time Series (Daily)"]
df = pd.DataFrame.from_dict(time_series, orient="index")
df = df.astype(float)
```

```python
df.index = pd.to_datetime(df.index)
df.sort_index(ascending=True, inplace=True)

```

```python

X = df.index.astype(int).values.reshape(-1, 1)
y = df["5. adjusted close"]

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

```

```python
# Print
future_dates = pd.date_range(start=df.index[-1] + pd.DateOffset(1), periods=5, freq="B")
future_dates_int = future_dates.astype(int).values.reshape(-1, 1)
predicted_prices = model.predict(future_dates_int)


for date, price in zip(future_dates, predicted_prices):
    print(f"Predicted price for {date.date()}: {price:.2f}")
```
